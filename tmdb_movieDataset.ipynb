# TMDB Movies Dataset 1923 to 2025 (upcoming included)

```python
import requests
import pandas as pd
import numpy as np
```

## **PAGE 1 DATA LOAD**

```python
response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1').json()
```

```python
popularity = []
vote_count = []
original_language =[]
title = []
vote_average = []
overview = []
release_date = []
original_title =[]
id = []
genre_ids = []

for i in response['results']:
      popularity.append(i['popularity'])
      vote_count.append(i['vote_count'])
      original_language.append(i['original_language'])
      title.append(i['title'])
      vote_average.append(i['vote_average'])
      overview.append(i['overview'])
      release_date.append(i['release_date'])
      original_title.append(i['original_title'])
      genre_ids.append(i['genre_ids'])
      id.append(i['id'])

d={'title':title ,"original_title" : original_title , 'id': id, 'genre_ids': genre_ids , 'vote_average' : vote_average , 'vote_count' : vote_count ,
   'original_language' : original_language , 'release_date' : release_date ,
   'popularity' : popularity , 'overview' : overview}
df = pd.DataFrame(d)
df
```

**TOTAL 500 PAGES LOAD**

```python
final = pd.DataFrame()

for j in range(1,501):

  response = requests.get('https://api.themoviedb.org/3/movie/popular?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(j)).json()

  popularity = []
  vote_count = []
  language =[]
  title = []
  vote_average = []
  overview = []
  release_date = []
  genre_ids = []

  for i in response['results']:
      popularity.append(i['popularity'])
      vote_count.append(i['vote_count'])
      language.append(i['original_language'])
      title.append(i['title'])
      vote_average.append(i['vote_average'])
      overview.append(i['overview'])
      release_date.append(i['release_date'])
      genre_ids.append(i['genre_ids'])


  d={'title':title , 'genres_ids' : genre_ids, 'vote_average' : vote_average , 'vote_count' : vote_count ,'original_language' : language ,
     'release_date' : release_date ,'popularity' : popularity , 'overview' : overview}

  df = pd.DataFrame(d)

  final = pd.concat([df,final] , ignore_index=True)

final
```

#Convert genres_ids into genres

```python
def replace_elements(lst, mapping):
    """
    Replace elements in a list using a mapping dictionary.

    :param lst: list of elements
    :param mapping: dict {old_value: new_value}
    :return: new list with replaced values
    """
    return [mapping.get(x, x) for x in lst]

for i in final['genres_ids']:

  mapping = {
      28: "action", 12:'adventure' , 16 :'animation' , 35 :'comedy',80:'crime', 99:'documentry',18:'drama',10751:'family' ,14:'fantasy',
      36:'history',27:'horror',10402:'music',9648:'mystery',878:'science fiction',10770:'TV movie',53:'thriller',10752:'war',37:'western'
    }
  genres = replace_elements(i, mapping)
  print(genres)
```
